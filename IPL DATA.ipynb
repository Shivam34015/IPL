```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#loading the ipl match dataset 
ipl=pd.read_csv('matches.csv')
```

```python
#having the glance of first five dataset of ipl data 
ipl.head()
```

```python
#looking at the number of rows and coloumns of dataset 
ipl.shape
```

```python
#getting the frequency of most man of the match awrdas
ipl['player_of_match'].value_counts()
```

```python
#getting the top10 players with most man of the match awrdas
ipl['player_of_match'].value_counts()[0:10]
```

```python
list(ipl['player_of_match'].value_counts()[0:10].keys())
```

```python
plt.figure(figsize=(8,5))
plt.bar (list(ipl['player_of_match'].value_counts()[0:5].keys()),list(ipl['player_of_match'].value_counts()[0:5]),color="g")
plt.show()
```

```python
#Extracting the reacords where a team won by batting first 
batting_first=ipl[ipl['win_by_runs']!=0]
```

```python
#Looking at the head
batting_first.head()
```

```python

#Making a histogram
plt.figure(figsize=(5,7))
plt.hist(batting_first['win_by_runs'])
plt.title("distribution of runs")
plt.xlabel("Runs")
plt.show()
```

```python
#finding out the nunmber of wins w.r.t each team after batting first 
batting_first['winner'].value_counts()
```

```python
#making a bar plot for top-3 teams with most wins after batting first 
plt.figure(figsize=(6,6))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),list(batting_first['winner'].value_counts()[0:3]),color=["blue","yellow","orange"])
plt.show()
```

```python
#making a piechart for batting first
plt.figure(figsize=(7,7))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
#extract those records where a team  has won after batting second
batting_second=ipl[ipl['win_by_wickets']!=0]
```

```python
#looking at the head
batting_second.head()
```

```python
plt.figure(figsize=(5,5))
plt.hist(batting_second['win_by_wickets'],bins=30)
plt.show()
```

```python
batting_second['winner'].value_counts()
```

```python
#making a barplot for top 3 teams with most win after batting second
plt.figure(figsize=(5,5))
plt.bar(list(batting_second['winner'].value_counts()[0:3].keys()),list(batting_second['winner'].value_counts()[0:3]),color=["purple","blue","red"])
plt.show()
```

```python
#making a piechart for distribution of most wins after batting second 
plt.figure(figsize=(5,5))
plt.pie(list(batting_second['winner'].value_counts()),labels=list(batting_second['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
